```python
# 在 Colab 新建单元格，保存为 optimize_batch.py

import os
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import torch
from net.CIDNet import CIDNet
from torchvision import transforms
import glob

def enhance_image(model, img_path, device, alpha_s=1.0, alpha_i=1.0, gamma=1.0):
    """单图推理"""
    img = Image.open(img_path).convert('RGB')

    # 预处理
    transform = transforms.Compose([
        transforms.ToTensor(),
    ])
    input_tensor = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(input_tensor ** gamma) ** (1/gamma)

    # 后处理
    output = output.squeeze(0).cpu().clamp(0, 1)
    output_img = transforms.ToPILImage()(output)

    return img, output_img

def create_comparison_grid(input_dir, output_dir, model, device, alpha_s=1.0, alpha_i=1.0, gamma=1.0):
    """批量生成对比图"""
    os.makedirs(output_dir, exist_ok=True)
    img_paths = sorted(glob.glob(f"{input_dir}/*.*"))

    for i, img_path in enumerate(img_paths):
        if i >= 9:  # 最多处理9张
            break

        try:
            input_img, output_img = enhance_image(model, img_path, device, alpha_s, alpha_i, gamma)

            # 拼接对比图
            fig, axes = plt.subplots(1, 2, figsize=(12, 5))
            axes[0].imshow(input_img)
            axes[0].set_title('Input (Low-light)', fontsize=14)
            axes[0].axis('off')

            axes[1].imshow(output_img)
            axes[1].set_title(f'Output (HVI-CIDNet)\nα_s={alpha_s}, α_i={alpha_i}, γ={gamma}', fontsize=14)
            axes[1].axis('off')

            plt.tight_layout()
            save_path = os.path.join(output_dir, f'compare_{i:02d}.png')
            plt.savefig(save_path, dpi=150, bbox_inches='tight')
            plt.close()

            print(f"✓ 处理完成: {os.path.basename(img_path)} -> {save_path}")

        except Exception as e:
            print(f"✗ 失败: {os.path.basename(img_path)} - {str(e)}")

def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"使用设备: {device}")

    # 加载模型
    model = CIDNet().to(device)

    # 加载权重（根据你下载的权重调整路径）
    checkpoint = torch.load('./weights/LOLv1/w_perc.pth', map_location=device)
    model.load_state_dict(checkpoint['state_dict'] if 'state_dict' in checkpoint else checkpoint)
    model.eval()

    # 批量处理
    create_comparison_grid(
        input_dir='./my_photos',      # 你的暗光照片文件夹
        output_dir='./output_compare', # 对比结果保存位置
        model=model,
        device=device,
        alpha_s=1.0,
        alpha_i=1.0,
        gamma=1.0
    )

    print("\n全部完成！对比图保存在 ./output_compare/")

if __name__ == '__main__':
    main()
```


# optimize_autoparam.py
import torch
import numpy as np
from PIL import Image
from torchvision import transforms
from net.CIDNet import CIDNet

def calculate_brightness(img_tensor):
    """计算图像亮度（RGB均值）"""
    return img_tensor.mean().item()

def calculate_contrast(img_tensor):
    """计算对比度（标准差）"""
    return img_tensor.std().item()

def auto_adjust_params(input_img_path, model, device):
    """根据输入图像亮度自适应调整参数"""
    img = Image.open(input_img_path).convert('RGB')
    transform = transforms.ToTensor()
    input_tensor = transform(img)
    
    brightness = calculate_brightness(input_tensor)
    contrast = calculate_contrast(input_tensor)
    
    print(f"输入图像亮度: {brightness:.3f}, 对比度: {contrast:.3f}")
    
    # 自适应规则（基于论文HVI色彩空间的特性）
    if brightness < 0.1:          # 极暗
        gamma = 0.5
        alpha_s = 1.2
        alpha_i = 1.5
    elif brightness < 0.3:        # 较暗
        gamma = 0.7
        alpha_s = 1.0
        alpha_i = 1.2
    elif brightness > 0.7:        # 偏亮
        gamma = 1.2
        alpha_s = 0.8
        alpha_i = 0.9
    else:                         # 正常
        gamma = 1.0
        alpha_s = 1.0
        alpha_i = 1.0
    
    print(f"自适应参数: gamma={gamma}, alpha_s={alpha_s}, alpha_i={alpha_i}")
    return gamma, alpha_s, alpha_i

def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = CIDNet().to(device)
    
    # 加载权重
    checkpoint = torch.load('./weights/LOLv1/w_perc.pth', map_location=device)
    model.load_state_dict(checkpoint['state_dict'] if 'state_dict' in checkpoint else checkpoint)
    model.eval()
    
    # 测试自适应参数
    img_path = './my_photos/dark_room.jpg'
    gamma, alpha_s, alpha_i = auto_adjust_params(img_path, model, device)
    
    # 用自适应参数推理
    img = Image.open(img_path).convert('RGB')
    transform = transforms.ToTensor()
    input_tensor = transform(img).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = model((input_tensor ** gamma) * alpha_s) ** (1/gamma) * alpha_i
    
    # 保存结果
    output = output.squeeze(0).cpu().clamp(0, 1)
    output_img = transforms.ToPILImage()(output)
    output_img.save('./output_adaptive.png')
    print("自适应结果已保存到 ./output_adaptive.png")

if __name__ == '__main__':
    main()
